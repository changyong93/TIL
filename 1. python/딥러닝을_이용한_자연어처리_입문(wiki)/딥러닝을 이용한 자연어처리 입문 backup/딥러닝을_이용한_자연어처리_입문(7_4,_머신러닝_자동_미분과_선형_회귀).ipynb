## 케라스로 구현하는 선형 회귀

- Sequential로 model이란 이름의 모델 만들기
-add를 통해 입력과 출력 벡터 차원과 같은 필요한 정보 추가

# 예시 코드. 실행 불가.
- model = keras.models.Sequential()
- model.add(keras.layers.Dense(1, input_dim=1))
---
- 첫 번째 인자 1 : 출력 차원 정의
- 두 번째 인자 unput_dim : 입력 차원 정의

```python
import numpy as np
from tensorflow.keras.models import Sequential
# Sequential 모델은 각 레이어에 정확히 하나의 입력 텐서와 하나의 출력 텐서가 있는 일반 레이어 스택에 적합

from tensorflow.keras.layers import Dense
# Dense : 신경망을 이용할 때 사용하는 모듈, Dense는 input을 넣었을 때 output으로 바꿔주는 중간 다리 역할
# https://han-py.tistory.com/207

from tensorflow.keras import optimizers
# optimizers : 최적화 함수를 설정하는 부분, 'sgd', 'adam' 'rmsprop'
```

image.png

- https://championprogram.tistory.com/273

```python
X = [1,2,3,4,5,6,7,8,9] # 공부하는 시간
y = [11,22,33,44,55,66,77,87,95] #각 공부하는 시간에 맵핑되는 성적
```

```python
model = Sequential()

#입력 x의 차원은 1, 출력 y의 차원도 1. 선형 회귀이므로 activation은 linear
model.add(Dense(1, activation='linear'))

# sgd는 경사하강법, 학습률(learning rate, lr)은 0.01
sgd = optimizers.SGD(lr = 0.01)

#손실 함수(Loss function)은 평균제옵오차 mse를 사용
model.compile(optimizer = sgd, loss = 'mse', metrics = ['mse'])

#주어진 X와 y 데이터에 대한 오차를 최소화 하는 작업 300번 시도
model.fit(X,y, batch_size = 1, epochs=300, shuffle = False)
```

```python
import matplotlib.pyplot as plt
plt.plot(X, model.predict(X), 'b', X,y, 'k.')
```
