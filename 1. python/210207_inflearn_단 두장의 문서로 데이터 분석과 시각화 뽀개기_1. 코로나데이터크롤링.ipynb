```python
def get_seoul_covid19_100(page_no):
    """
    page_no : 입력값으로 페이지 번호를 입력하면 해당 페이지의 데이터를 가져옴
    start_no : 입력받은 page_no로 시작하는 번호
    """
    start_no = (page_no-1)*100
    url = f'https://news.seoul.go.kr/api/27/getCorona19Status/get_status_ajax.php?draw={page_no}'
    url = f'{url}&start={start_no}&length=100'
    
    response = requests.get(url)
    data_json = response.json()
    return data_json
```

```python
page = round(get_seoul_covid19_100(1)['recordsTotal']/100)
```

```python
get_seoul_covid19_100(page+1)
```

```python
data_json = get_seoul_covid19_100(1)
pd.DataFrame(data_json['data'])
```

```python
import time #time.sleep으로 시간간격을 두기 위해
from tqdm import trange #진행 상태를 표시하기 위해
```

```python
def get_multi_page_data(start_page, end_page = 80):
    all_page = round(get_seoul_covid19_100(1)['recordsTotal']/100)
    covid19_data = []
    
    for page_no in trange(start_page,end_page+1):
        data_json = get_seoul_covid19_100(page_no)

        if len(data_json['data'])>0:
            data_df = pd.DataFrame(data_json['data'])
            covid19_data.append(data_df)
            time.sleep(0.5)
        else:
            return covid19_data
    
    covid19_data = pd.concat(covid19_data)
    print(f'총 페이지 수 : {all_page}')
    return covid19_data
```

```python
start_page = 1
end_page = 152
page_list = get_multi_page_data(1,152)
```

```python
page_list.columns = col_list
```

```python
strings = page_list.iloc[0,0]
strings
```

```python
import re #정규표현식 사용을 위해 re library 설정
def extract_number(num_string):
    if type(num_string) == str:
        num_string = num_string.replace("corona19","")
        num = re.sub(pattern = "[^0-9]",repl = "", string = num_string)
        num = int(num)
        return num
    else:
        return num_string
```

```python
page_list["연번"] = page_list["연번"].map(extract_number)
page_list
```

```python
def extract_hangeul(origin_text):
    substract_text = re.sub("[^가-힣]","",origin_text)
    return substract_text
```

```python
page_list["퇴원현황"] = page_list["퇴원현황"].map(extract_hangeul)
page_list
```

```python
page_list["퇴원현황"].value_counts()
```

```python
page_list.loc[page_list["퇴원현황"].str.contains("퇴원"),"퇴원현황"] = "퇴원"
page_list.loc[page_list["퇴원현황"].str.contains("사망"),"퇴원현황"] = "사망"
page_list.loc[~page_list["퇴원현황"].str.contains("퇴원|사망"),"퇴원현황"] = np.nan
page_list
```
